```python
import pandas as pd
```

```python
# Reading the data file and assigning it to variable df
df= pd.read_csv("Excl_exp3.csv")
df.head()
```

```python
# Removal of unwanted column and identifying the right index
df = df.set_index('Unnamed: 0')
```

```python
df.head()
```

```python
# Calling the matplotlib library to use its functions for plotting
import matplotlib.pyplot as plt
```

```python
# Calling sklearn library to use its functions for the calculation of Sum of Squared distances to draw the Elbow graph
from sklearn.cluster import KMeans
Sum_of_squared_distances = []
# K is the range of the clusters number that is going to be introduced to a loop to calculate the Sum of Squared distances
K = range(1,50)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
# Drawing the Elbow graph using the different results of the Sum of Squared distances against their numbers of clusters
plt.figure(figsize=(15,7))
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum of squared distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

### As shown above, We can recognize the curve Elbow (breaking point) at k = 26, which means that the expected optimal number of clusters is 26.
